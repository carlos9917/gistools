## Quick visualization of the stations


```python
import warnings
warnings.filterwarnings('ignore')
import os
import geopandas as gpd
import folium as fl
import pandas as pd

DPATH="~/sync_laptops/NOTES/RoadWeather/"
cols=["SID","c2","c3","SID_again","name","lon","lat","creation_date"]+["c"+str(i+8) for i in range(15)]

added_202301 = pd.read_csv(os.path.join(DPATH,"stations_added_202301.csv"),header=None,names=cols) 
added_202302 = pd.read_csv(os.path.join(DPATH,"stations_added_202302.csv"),header=None,names=cols) 
added_202303 = pd.read_csv(os.path.join(DPATH,"stations_added_202303.csv"),header=None,names=cols) 

added_202201 = pd.read_csv(os.path.join(DPATH,"stations_added_202201.csv"),header=None,names=cols) 
added_202202 = pd.read_csv(os.path.join(DPATH,"stations_added_202202.csv"),header=None,names=cols) 
added_202203 = pd.read_csv(os.path.join(DPATH,"stations_added_202203.csv"),header=None,names=cols) 


#There were no new stations added in the months of Jan-Mar in 2021

#added_202101 = pd.read_csv(os.path.join(DPATH,"stations_added_202101.csv"),header=None) 
#added_202102 = pd.read_csv(os.path.join(DPATH,"stations_added_202102.csv"),header=None) 
#added_202103 = pd.read_csv(os.path.join(DPATH,"stations_added_202103.csv"),header=None) 


#added_202301.columns = cols
#added_202302.columns

#save these as geopandas for later
geo_202201 = gpd.GeoDataFrame(
    added_202201, geometry=gpd.points_from_xy(added_202201.lon, added_202201.lat), crs="EPSG:4326"
)
geo_202201.to_file("stations_added_202201.shp")

geo_202202 = gpd.GeoDataFrame(
    added_202202, geometry=gpd.points_from_xy(added_202202.lon, added_202202.lat), crs="EPSG:4326"
)
geo_202202.to_file("stations_added_202202.shp")

geo_202203 = gpd.GeoDataFrame(
    added_202203, geometry=gpd.points_from_xy(added_202203.lon, added_202203.lat), crs="EPSG:4326"
)
geo_202203.to_file("stations_added_202203.shp")

geo_202301 = gpd.GeoDataFrame(
    added_202301, geometry=gpd.points_from_xy(added_202301.lon, added_202301.lat), crs="EPSG:4326"
)
geo_202201.to_file("stations_added_202301.shp")

geo_202302 = gpd.GeoDataFrame(
    added_202302, geometry=gpd.points_from_xy(added_202302.lon, added_202302.lat), crs="EPSG:4326"
)
geo_202302.to_file("stations_added_202302.shp")

geo_202303 = gpd.GeoDataFrame(
    added_202303, geometry=gpd.points_from_xy(added_202303.lon, added_202303.lat), crs="EPSG:4326"
)
geo_202303.to_file("stations_added_202303.shp")

enc='cp865' # 'utf-8'
all_stations = pd.read_csv(os.path.join(DPATH,"vejvejr_long_names_identify_new_CLEANED.csv"),encoding=enc,header=None,names=cols) 

geo_all_stations = gpd.GeoDataFrame(all_stations, 
    geometry=gpd.points_from_xy(all_stations.lon, all_stations.lat), crs="EPSG:4326"
)

geo_all_stations.to_file("all_vejvejr_stations_2023.shp")
                                    
    
```

### Plot the stations added in each month over each year



```python
tooltip = "Click Here For Station name"


loc = 'Stations added in January. Red for 2022, Blue for 2023'
title_html = '''
             <h3 align="center" style="font-size:16px"><b>{}</b></h3>
             '''.format(loc)   






m = fl.Map(zoom_start=50, layers_control=True, tiles="OpenStreetMap")
for _, r in added_202201.iterrows():
    #print(r)
    station = r["SID"]
    this_lat = r["lat"]
    this_lon = r["lon"]
    icon=fl.Icon(color='red')
    #l2rc_group.add_child(folium.Marker((df[i, 'Latitude'],df[i, 'Longitude']), icon=icon))
    marker = fl.Marker(
        location=[this_lat, this_lon],
        popup=f"<stong>Added Jan 2022: {station}</stong>",
        tooltip=tooltip, icon=icon)
    marker.add_to(m)


for _, r in added_202301.iterrows():
    #print(r)
    station = r["SID"]
    this_lat = r["lat"]
    this_lon = r["lon"]
    icon=fl.Icon(color='blue')
    #l2rc_group.add_child(folium.Marker((df[i, 'Latitude'],df[i, 'Longitude']), icon=icon))
    marker = fl.Marker(
        location=[this_lat, this_lon],
        popup=f"<stong>Added Jan 2023: {station}</stong>",
        tooltip=tooltip, icon=icon)
    marker.add_to(m)
    
m.get_root().html.add_child(fl.Element(title_html))
    
m
m.save('map.html')
```
